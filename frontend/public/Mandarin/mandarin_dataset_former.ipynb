```python
with open("./eng_src.txt", "r", encoding='utf-8') as f:
    eng_lines = f.readlines()

with open("./flores_mandarin_mt.txt", "r", encoding='utf-8') as f:
    mt_lines = f.readlines()

with open("./mandarin_ref.txt", "r", encoding='utf-8') as f:
    ref_lines = f.readlines()
```

```python
count = 1
newSet = []
for ind,line in enumerate(eng_lines):
    # if count == 5:
        # break
    line = line.replace('"', "'")
    mt_line = mt_lines[ind].replace('"', "'")
    ref_line = ref_lines[ind].replace('"', "'")
    newLine = f'{{"id": {count}, "src": "{line.strip()}", "mt": "{mt_line.strip()}", "ref": "{ref_line.strip()}"}},'
    newSet.append(newLine)
    # print(newLine)
    count += 1
```

```python
newSet

with open("./mandarin_dataset.json", "w", encoding='utf-8') as f:
    f.write("[\n")
    for line in newSet:
        f.write(line + "\n")
    f.write("]")
```
